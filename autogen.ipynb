```python
from IPython.display import display, HTML
from autogen_agentchat.agents import AssistantAgent
from autogen_core.models import UserMessage
from autogen_ext.models.openai import AzureOpenAIChatCompletionClient

from azure.core.credentials import AzureKeyCredential
from autogen_core import CancellationToken

from autogen_agentchat.messages import TextMessage
from autogen_agentchat.ui import Console
import asyncio
import markdown


from dotenv import load_dotenv
import os

load_dotenv()



model_client = AzureOpenAIChatCompletionClient(
    azure_deployment=os.environ.get("AZURE_OPENAI_LLM", ""),
    model="gpt-4o",
    api_version=os.environ.get("OPENAI_API_VERSION", ""),
    azure_endpoint=os.environ.get("AZURE_OPENAI_ENDPOINT", ""),
    api_key=os.environ.get("AZURE_OPENAI_API_KEY", "")
)

agent = AssistantAgent(
    name="assistant",
    model_client=model_client,
    tools=[],
    system_message="You are a travel agent that plans great vacations",
)

async def assistant_run():
    from IPython.display import HTML, display
    import markdown

    user_query = "Plan me a great sunny vacation"

    # Get agent response
    response = await agent.on_messages(
        [TextMessage(content=user_query, source="user")],
        cancellation_token=CancellationToken(),
    )

    # Convert Markdown to HTML
    markdown_html = markdown.markdown(response.chat_message.content)

    # Create styled output
    full_html = f"""
    <div style="font-family: Arial; line-height: 1.6; max-width: 800px">
        <div style="font-weight: bold; margin-bottom: 5px">User:</div>
        <div style="margin-left: 20px; background: #f0f8ff; padding: 10px; border-radius: 4px">
            {user_query}
        </div>
        <div style="font-weight: bold; margin-top: 20px; margin-bottom: 5px">Assistant:</div>
        <div style="margin-left: 20px; background: #f5f5f5; padding: 10px; border-radius: 4px">
            {markdown_html}
        </div>
    </div>
    """

    display(HTML(full_html))
```

```python
# Run the function
await assistant_run()
```
